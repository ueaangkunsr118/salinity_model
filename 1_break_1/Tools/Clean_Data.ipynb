```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Path to your CSV file
file_path = '/Users/ratchanonkhongsawi/Desktop/CMKL/SELF-Learn/RND1/Data/chaophraya2567 (1).csv'  # Replace 'data.csv' with the path to your file

# Load the data into a DataFrame
df = pd.read_csv(file_path)

# Get unique station names
unique_station_names = df['stn_name'].unique()
```

```python
csv_file1 = 'การท่าเรือ.csv'
df1 = pd.read_csv(csv_file1)
# Convert datetim to datetimes format and set as the index
df1['datetimes'] = pd.to_datetime(df1['datetimes'], format='%Y-%m-%d %H:%M:%S')
df1['datetimes_copy'] = df1['datetimes']
df1.set_index('datetimes', inplace=True)

# Filter for columns 'Salinity' and 'Sensor_Depth' and remove rows with NaN values
# df1 = df1.loc[: , ['Salinity']]

# Plot both columns
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Salinity data
ax1.plot(df1.index, df1['salinity'], '.', markersize=0.5, label='salinity', color='b')
ax1.set_xlabel('Date')
ax1.set_ylabel('salinity', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Title and show plot
plt.title('Salinity, df1')
plt.show()
```
